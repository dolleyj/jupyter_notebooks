```python
import pandas as pd
import requests

# Being deprecated June 1, 2019
# https://iextrading.com/developer/docs/#key-stats
# Import: https://api.iextrading.com/1.0/stock/market/batch?symbols=aapl,fb&types=stats&range=1m&last=5

# New API - https://iexcloud.io/console/

iexcloud_secret = 'sk_30bb73ac8fb44029b42aa19d64033d00'
iexcloud_publishable = 'pk_af65282ca87d4130a74e810a1a767360'

stocks = ['CTT', 'CONE', 'DIS', 'FRT', 'IRM', 'KIM', 'PG', 'O', 'STOR', 'SKT', 'VTR', 'VZ']
stocks_string = (',').join(stocks)
# url = "https://api.iextrading.com/1.0/stock/market/batch?symbols="+ (',').join(stocks).lower() +"&types=stats,financials&range=1m&last=5"
url = 'https://cloud.iexapis.com/stable/stock/market/batch?symbols='+stocks_string+'&types=stats,financials&token='+iexcloud_publishable
print(url)

r = requests.get(url)
data_json = r.json()

# JSON nests stats. This moves them up and out of that nest
# {'FB':{'stats':{stats_dict},'AAPL':{...}} => {'FB':{stats_dict},'AAPL':{...}}
stats_by_company = {}
financials_by_company = {}
for comp in data_json.keys():
    stats_by_company[comp] = data_json[comp]['stats']
    financials_by_company[comp] = data_json[comp]['financials']['financials'][0]
    
df_stats = pd.DataFrame(stats_by_company).transpose()
df_financials = pd.DataFrame(financials_by_company).transpose()
```
